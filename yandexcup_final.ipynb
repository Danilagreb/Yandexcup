```python
"""
Задание: дан датасет с данными по пополнению/снятию бонусов у определенного сервиса.
Необходимо вывести балансы сервисов по пользователям, где было хотя бы одно пополнение.

Решение:
Были отобраны столбцы, где было хотя бы одно пополнение. Выполнена группировка для
выявления списка нужных столбцов. Затем отобрали нужные значения строк и просуммировали
по ним. Получили балансы у сервисов по определенным пользователям и сервисам, где были
пополнения бонусов. Ответ представлен в DataFrame da
"""
```

```python

data = pd.read_csv('points_task_input.csv') # чтение данных

data_1 = data.copy()
data_1 = data_1[data_1['amount'] > 0] # отбираем данные, в которых точно встретится пополнение
data_1 = data_1.groupby(['id', 'service'], as_index=False).sum() # группируем для счета баланса
data_1.drop('amount', axis=1, inplace=True)
data_1['id_service'] = data_1['id'].astype(str) + '_' + data_1['service'] # объединение двух столбцов для уникальности
data_1.drop(['id', 'service'], axis=1, inplace=True)


data_test = data.copy()
data_test['id_service'] = data_test['id'].astype(str) + '_' + data_test['service'] # объединение двух столбцов для уникальности
data_test.drop(['id', 'service'], axis=1, inplace=True)
list_serv = data_1['id_service'].unique() # берем уникальные значения
data_test = data_test.groupby('id_service', as_index=False).sum() # группировка с применением суммы для подсчета
da = data_test[data_test['id_service'].isin(list_serv)]

da[['id','service']] = da['id_service'].str.split('_',expand=True)
da.drop('id_service', axis=1, inplace=True)
da = da[['id', 'service', 'amount']]
da = da.rename(columns={'amount': 'balance'}) # переименуем столбец, так как в нем отражен баланс
da = da.sort_values(by=['id', 'service'])
da # получаем баланс пользователей для каждого сервиса, где были пополнения
```
